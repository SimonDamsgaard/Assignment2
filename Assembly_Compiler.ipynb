```python
#Compileren


bruv = []


def NumtoBinStrBit(num, bits):
    binary_string = format(num, f"0{bits}b")
    return binary_string

def RegtoStrBit(reg, bits):
    num = int(reg[1])
    return NumtoBinStrBit(num, bits)


def AddZeros(binstr):
    brotha = binstr
    while len(brotha) < 32:
        brotha += "0"
    return brotha
        



with open("C:/Users/hanst/Desktop/DTU/3 semester/Computersystemer/Aflevering2/Assembly_Without_Comments.txt", "r") as file:
    # content = file.read()       # Reads entire file
    # #print(content)
    # for i in content:
    #     print(i)
    for line in file:
        bruv.append(line)




# Writing the read data to another file
with open("C:/Users/hanst/Desktop/DTU/3 semester/Computersystemer/Aflevering2/output.txt", "w") as outfile:
    op_codes = {
        "ADD": "100000",
        "ADDI": "110001",
        "SUBI": "110010",
        "MULI": "110011",
        "LI": "010000",
        "LD": "001001",
        "SD": "001010",
        "BEQ": "100101",
        "END": "000000"
    }
    for i in bruv:

        sis = i.split()

        op = sis[0]

        sendstring = ""

        sendstring += op_codes.get(op)

        if op == "ADD":
            sumreg = sis[1]
            addreg_1 = sis[2]
            addreg_2 = sis[3]

            sendstring += RegtoStrBit(sumreg, 4)
            sendstring += RegtoStrBit(addreg_1, 4)
            sendstring += RegtoStrBit(addreg_2, 4)
        
        if op == "ADDI":
            sumreg = sis[1]
            addreg = sis[2]
            addint = int(sis[3])

            sendstring += RegtoStrBit(sumreg, 4)
            sendstring += RegtoStrBit(addreg, 4)
            sendstring += NumtoBinStrBit(addint, 9)

        if op == "SUBI":
            diffreg = sis[1]
            startreg = sis[2]
            subint = int(sis[3])

            sendstring += RegtoStrBit(diffreg, 4)
            sendstring += RegtoStrBit(startreg, 4)
            sendstring += NumtoBinStrBit(subint, 9)

        if op == "MULI":
            proreg = sis[1]
            facreg = sis[2]
            facint = int(sis[3])

            sendstring += RegtoStrBit(proreg, 4)
            sendstring += RegtoStrBit(facreg, 4)
            sendstring += NumtoBinStrBit(facint, 9)
        
        if op == "LI":
            tarreg = sis[1]
            tarint = int(sis[2])

            sendstring += RegtoStrBit(tarreg, 4)
            sendstring += NumtoBinStrBit(tarint, 9)

        if op == "LD":
            tarreg = sis[1]
            memaddreg = sis[2]

            sendstring += RegtoStrBit(tarreg, 4)
            sendstring += RegtoStrBit(memaddreg, 4)
        
        if op == "SD":
            valreg = sis[1]
            memaddreg = sis[2]

            sendstring += RegtoStrBit(valreg, 4)
            sendstring += RegtoStrBit(memaddreg, 4)
        
        if op == "BEQ":
            linum = int(sis[1])
            comreg_1 = sis[2]
            comreg_2 = sis[3]

            sendstring += NumtoBinStrBit(linum, 6)
            sendstring += RegtoStrBit(comreg_1, 4)
            sendstring += RegtoStrBit(comreg_2, 4)
          
        
        yallah = AddZeros(sendstring)
        sendstring = yallah
        print(sendstring)
        outfile.write(str(sendstring) + "\n")
```
